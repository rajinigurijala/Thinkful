```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
df = pd.read_csv('titanic.csv')
#fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(2, 1)

#df.groupby('Survived')['PassengerId'].count().sort_values().plot(kind='bar', ax=ax1)
df1=df.groupby(('Survived')).size()*100 / len(df)
df1.plot(kind='barh',ax=ax1)
ax1.yaxis.set_label_text("Survival %")
ax1.set_title("Survival Rate")
#use the tick functionality to convert the labels


# Divide the figure into a 1x1 grid, and give me the section
df.groupby('Sex')['Survived'].sum().sort_values().plot(kind='bar', ax=ax2)
ax2.yaxis.set_label_text("Survived")
ax2.set_title("Survived by Gender")
fig.subplots_adjust(hspace=0.5)
```

```python
survived = df.loc[df['Survived'] == 1].count()[0]
not_survived = df.loc[df['Survived'] == 0].count()[0]
print(survived)
print(not_survived)
```

```python
survived = df.loc[df['Survived'] == 1].count()[0]
not_survived = df.loc[df['Survived'] == 0].count()[0]

plt.figure(figsize=(8,5))

labels = ['Survived', 'Not Survived']

plt.pie([survived, not_survived], labels = labels,  autopct='%.2f %%')

plt.title('Survival Rate')

plt.show()
```

```python
survived = 'survived'
not_survived = 'not survived'
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(10, 4))
women = df[df['Sex']=='female']
ax1.hist(women[women['Survived']==1].Age.dropna(), bins=18, label = survived, color = '#abcdef')
ax1.hist(women[women['Survived']==0].Age.dropna(), bins=40, label = not_survived, color = 'Orange')
ax1.set_title('Female')
ax1.xaxis.set_label_text("Age")
ax1.legend()

men = df[df['Sex']=='male']
ax2.hist(men[men['Survived']==1].Age.dropna(), bins=18, label = survived, color = '#abcdef')
ax2.hist(men[men['Survived']==0].Age.dropna(), bins=40, label = not_survived, color = 'Orange')
ax2.set_title('Male')
ax1.xaxis.set_label_text("Age")
ax2.legend()

plt.show()
```
